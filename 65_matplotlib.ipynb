```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[5,7,4])
plt.show()
```

```python
plt.plot([1,2,3],[5,7,4])
plt.title("student chart")
plt.xlabel('markx')
plt.ylabel('marky')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x=[4,3,2]
y=[8,2,3]
x2=[1,2,6]
y2=[5,4,3]
plt.plot(x,y,label='first',linewidth=3)
plt.plot(x2,y2, label='second')
plt.legend() #to use label this function should must be called
plt.show()
```

```python
x=[4,3,2]
y=[8,2,3]
x2=[1,2,6]
y2=[5,4,3]
plt.scatter(x,y)
plt.plot(x2,y2)
```

```python
x=[4,3,2]
y=[8,2,3]
x2=[1,2,6]
y2=[5,4,3]
plt.hist(x) #will create histogram;sequence of (n,);integer are sequence or auto opion
plt.plot(x2,y2)
```

```python

plt.figure(figsize=(3,3))
x=[4,3,2]
y=[8,2,3]
x2=[1,2,6]
y2=[5,4,3]
plt.barh(y,x)
plt.barh(y2,x2,color='green')
plt.show()
```

```python
x=[4,3,2,45,27,14,38]
y=[8,2,3,78,56,12,53]
x2=[1,2,6]
y2=[5,4,3]
#plt.subplot(no. of rows,no. of cols,no. of index)
plt.subplot(4,3,5)
plt.barh(y,x)
plt.show()
plt.subplot(5,2,3)
plt.barh(y2,x2,color='green')
plt.show()
```
